```python
import pandas as pd
```

```python
df = pd.read_csv('https://query.data.world/s/442hmws6ei6qe3xfgbp6tdemtxxz7p')
df.head()
```

```python
#CONTIONOUS DATA PLOT
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#histogram
plt.figure(figsize=(5,2))
sns.displot(df['tenure'])
```

```python
#density plot
sns.displot(df['tenure'], kind = 'kde', height = 2, aspect = 2)
```

```python
#box plot
plt.figure(figsize=(5,2))
sns.boxplot(x = df['tenure'])
```

```python
#box plot
plt.figure(figsize=(5,2))
sns.boxplot(x = df['MonthlyCharges'])
```

```python
#box plot
plt.figure(figsize=(5,2))
sns.boxplot(x = df[df['Churn'] == 'Yes' ]['MonthlyCharges'])
```

```python
#box plot
plt.figure(figsize=(5,2))
sns.boxplot(x = df[df['Churn'] == 'No' ]['MonthlyCharges'])
```

```python
# multiple plots in one figure
plt.figure(figsize=(10,2))

plt.subplot(1,2,1)
plt.title("Churn MonthlyCharges")
sns.boxplot(x = df[df['Churn'] == 'Yes' ]['MonthlyCharges'])

plt.subplot(1,2,2)
plt.title("Non-Churn MonthlyCharges")
sns.boxplot(x = df[df['Churn'] == 'No' ]['MonthlyCharges'])
```

```python
# two figures in one scale

sns.boxplot(x=df['Churn'], y =df['MonthlyCharges'])
```
