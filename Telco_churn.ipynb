```python
# by suing groupby, we can perform grouping on only row level
df.groupby(['Contract', 'Churn']).agg({'MonthlyCharges' : 'mean'})
```

```python
def t_b(x):
  if x<5:
    return 'D'
  elif x>=5 and x<20:
    return 'C'
  elif x>= 20 and x<50:
    return 'B'
  else:
    return 'A'
```

```python
df['tenure_b'] = df['tenure'].apply(t_b)
```
